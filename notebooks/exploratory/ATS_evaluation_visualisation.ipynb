```python
df_pbp['fixed_drive_result_points'] = df_pbp['fixed_drive_result'].map({'Punt':0,
                                                                        'Touchdown':7, 
                                                                        'Turnover':0, 
                                                                        'Field goal':3, 
                                                                        'Missed field goal':0,
                                                                        'Opp touchdown':-7, 
                                                                        'End of half':0, 
                                                                        'Safety':-2, 
                                                                        None:0,
                                                                        'Turnover on downs':0})
```

```python
temp = df_pbp.loc[:,['down','ydstogo','yardline_100', 'posteam_type','fixed_drive_result', 'fixed_drive_result_points','ep']].dropna()
temp.sort_values(['down','ydstogo','yardline_100', 'posteam_type'])
```
