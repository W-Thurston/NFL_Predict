```python
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('NFL_wk_by_wk.csv', names=['WEEK_NUM','GAME_DAY_OF_WEEK','GAME_DATE','GAMETIME','WINNER','GAME_LOCATION','LOSER','BOXSCORE_LINK','PTS_WINNER','PTS_LOSER','YARDS_WINNER','TURNOVERS_WINNER','YARDS_LOSER','TURNOVERS_LOSER'], index_col=False)
df = df[((df['WEEK_NUM']!='Week')&(df['WEEK_NUM']!='NULL_VALUE'))]
df['WINNER'].replace({'St. Louis Rams':'Los Angeles Rams',
                          'San Diego Chargers':'Los Angeles Chargers'},inplace=True)
df['LOSER'].replace({'St. Louis Rams':'Los Angeles Rams',
                          'San Diego Chargers':'Los Angeles Chargers'},inplace=True)
df.head()
```

```python
def elo_win_probability(rating_teamA, rating_teamB):
    P1 = 1.0 / (1.0 + 10**((rating_teamB-rating_teamA)/400))
    P2 = 1.0 / (1.0 + 10**((rating_teamA-rating_teamB)/400))
    
    print(f"Probability of Team A winning: {P1:.1%}")
    print(f"Probability of Team B winning: {P2:.1%}")
```

```python
elo_win_probability(1200, 1000)
```

```python
Suppose there is a live match on chess.com between two players
rating1 = 1200, rating2 = 1000;
P1 = (1.0 / (1.0 + pow(10, ((1000-1200) / 400)))) = 0.76
P2 = (1.0 / (1.0 + pow(10, ((1200-1000) / 400)))) = 0.24

And Assume constant K=30;

CASE-1 : Suppose Player 1 wins:
rating1 = rating1 + k*(actual – expected) = 1200+30(1 – 0.76) = 1207.2;
rating2 = rating2 + k*(actual – expected) = 1000+30(0 – 0.24) = 992.8;

Case-2 : Suppose Player 2 wins:
rating1 = rating1 + k*(actual – expected) = 1200+30(0 – 0.76) = 1177.2;
rating2 = rating2 + k*(actual – expected) = 1000+30(1 – 0.24) = 1022.8; 
```

```python
df = pd.DataFrame()
for i in range(2010,2019):
    temp = pd.concat(pd.read_html(f'https://www.pro-football-reference.com/years/{i}/games.htm', header=0))
    temp = temp[temp['Week']!='Week']
    temp['Year'] = str(i)[-2:]+'-'+str(int(str(i)[-2:])+1)
    temp.rename({'Unnamed: 5':'Game Location'},axis=1,inplace=True)
    temp.drop(['Unnamed: 7'],axis=1, inplace=True)
    print(temp.shape)
    
    df = pd.concat([df, temp])
    
df['Winner/tie'].replace({'St. Louis Rams':'Los Angeles Rams',
                          'San Diego Chargers':'Los Angeles Chargers'},inplace=True)
print(df.shape)
df.head()
```
